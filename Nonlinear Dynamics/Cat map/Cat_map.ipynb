```python
from PIL import Image 
from numpy import* 
from matplotlib import pyplot as plt
import numpy as np
import copy
```

```python
temp=asarray(Image.open('face.jpg')) 
plt.imshow(temp)

size = temp.shape[:2]
print(size[0])
```

```python
N = size[0]
x_next = 0
y_next = 0
steps_num = 300
step = 0
temp2 = copy.copy(temp)

while step < steps_num:
  for x in range(N):
    for y in range(N):
      x_next = (2*x + y) % N
      y_next = (x + y) % N
      temp2[x_next][y_next] = temp[x][y]

  #plt.imshow(temp)

  name = 'Output1/' + str(step) + '.jpg'
  im = Image.fromarray(temp)
  im.save(name, 'JPEG')

  temp = copy.copy(temp2)

  step += 1
```
